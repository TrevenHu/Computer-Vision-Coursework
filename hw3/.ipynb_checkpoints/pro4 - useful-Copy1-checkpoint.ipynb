```python
import numpy as np
from scipy.misc import imread
import matplotlib.pyplot as plt
from scipy.ndimage.filters import gaussian_filter
from numpy import linalg as la
```

```python
def rgb2gray(rgb):
    """ Convert rgb image to grayscale.
    """
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
# read images and detect corners on images
imgs_mat = []
crns_mat = []
imgs_war = []
crns_war = []
img_mat = imread('p4/matrix/matrix' + str(1) + '.png')
imgs_mat = rgb2gray(img_mat)
```

```python
dx = np.array([[ -1,  0,  1],[ -1,  0,  1],[-1,0,1]])
dy = dx.T
```

```python
from scipy import signal
from scipy import misc
Ix = signal.convolve2d(imgs_mat, dx, mode='valid')
Iy = signal.convolve2d(imgs_mat, dy, mode='valid')
```

```python
windowSize = 3
r = (windowSize-1)/2
dx, dy = np.meshgrid(np.arange(-r,r+1),np.arange(-r,r+1))
```

```python
def gs(s,x,y):
    return np.exp(-((x*x) + (y*y))/SmoothSTD/s*s)
```

```python
SmoothSTD = 2
weight = gs(SmoothSTD, dx, dy)/sum(sum(gs(SmoothSTD, dx, dy)))
```

```python
Ixx = signal.convolve2d(Ix*Ix, weight, mode='valid')
Iyy = signal.convolve2d(Iy*Iy, weight, mode='valid')
Ixy = signal.convolve2d(Ix*Iy, weight, mode='valid')
```

```python
plt.imshow(np.sqrt(Iyy+Ixx),cmap='gray')
plt.show()
```

```python
def findMax(n,mm):
    tmp = mm.copy()
    for i in range(0,n):
        maxi = 1
        xx = 0
        yy = 0
        for x in range(0,tmp.shape[0]):
            for y in range(0,tmp.shape[1]):
                if tmp[x,y] > maxi:
                    maxi = tmp[x,y]
                    xx = x
                    yy = y
        tmp[xx,yy] = 0
        corners[i,0] = yy
        corners[i,1] = xx
```

```python
nCorners = 20
corners = np.zeros((nCorners, 2))
p = localmin1.copy()
findMax(nCorners,p)
```
