## Problem 3: Normalized Cross Correlation (NCC) [3 pts]

Hint: Express $c_{NCC}$ and $c_{NSSD}$ in terms of the vectors $\tilde{w_1} =  \tilde{W_1} (:)$
and $\tilde{w_2} =  \tilde{W_2} (:)$. Thus, $c_{NCC} = \tilde{w_1}^T\tilde{w_2}$ and $c_{NSSD} = (\tilde{w_1} - \tilde{w_2})^T(\tilde{w_1} - \tilde{w_2})$,

$\tilde{W} = \frac{W - \overline{W}}{\sqrt{\sum_{k,l}(W(k,l) - \overline{W})^2}}$ is a mean-shifted and normalized version of the window.

$c_{NSSD} = (\tilde{w_1} - \tilde{w_2})^T(\tilde{w_1} - \tilde{w_2})\\

```python
import numpy as np
from scipy.misc import imread
import matplotlib.pyplot as plt
from scipy.ndimage.filters import gaussian_filter
```

```python
def rgb2gray(rgb):
    """ Convert rgb image to grayscale.
    """
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
from scipy import signal
from scipy import misc
def gs(s,x,y):
    return np.exp(-((x*x) + (y*y))/2/s*s)
def findMax(n,mm,c):
    tmp = mm.copy()
    for i in range(0,n):
        maxi = 1
        xx = 0
        yy = 0
        for x in range(0,tmp.shape[0]):
            for y in range(0,tmp.shape[1]):
                if tmp[x,y] > maxi:
                    maxi = tmp[x,y]
                    xx = x
                    yy = y
        tmp[xx,yy] = 0
        c[i,0] = yy
        c[i,1] = xx
def corner_detect(image, nCorners, smoothSTD, windowSize):
    """Detect corners on a given image.

    Args:
        image: Given a grayscale image on which to detect corners.
        nCorners: Total number of corners to be extracted.
        smoothSTD: Standard deviation of the Gaussian smoothing kernel.
        windowSize: Window size for corner detector and non maximum suppression.

    Returns:
        Detected corners (in image coordinate) in a numpy array (n*2).

    """
    dx = np.array([[ -1,  0,  1],[ -1,  0,  1],[-1,0,1]])
    dy = dx.T
    
    Ix = signal.convolve2d(image, dx, mode='valid')
    Iy = signal.convolve2d(image, dy, mode='valid')
    
    r = (windowSize-1)/2
    dx, dy = np.meshgrid(np.arange(-r,r+1),np.arange(-r,r+1))
    w = gs(smoothSTD, dx, dy)/sum(sum(gs(smoothSTD, dx, dy)))
    
    Ixx = signal.convolve2d(Ix*Ix, w, mode='valid')
    Iyy = signal.convolve2d(Iy*Iy, w, mode='valid')
    Ixy = signal.convolve2d(Ix*Iy, w, mode='valid')
    
    size = image.shape
    q1 = np.zeros((size[0]*size[1],1))
    q2 = np.zeros((size[0]*size[1],1))
    q3 = np.zeros((size[0]*size[1],1))
    
    s = np.ones((windowSize,windowSize))
    q1= signal.convolve2d(Ixx, s, mode='same')
    q2= signal.convolve2d(Iyy, s, mode='same')
    q3= signal.convolve2d(Ixy, s, mode='same')
    
    t = (q1+q2)*(q1+q2)
    tt = 4*(q1*q2-q3*q3)
    k = abs(t-tt)
    S = np.sqrt(k)
    
    lam1 = (q1+q2+S)/2
    lam2 = (q1+q2-S)/2
    lam = np.zeros((np.shape(lam1)[0],np.shape(lam1)[1],2))
    lam[:,:,0] = lam1
    lam[:,:,1] = lam2
    
    sz = lam1.shape
    lam = np.reshape(lam,(np.shape(lam)[0]*np.shape(lam)[1], np.shape(lam)[2]))
    
    localmin = np.min(lam,axis=1)
    localmin = np.reshape(localmin,(sz[0], sz[1]))
    row = np.shape(localmin)[0]
    col = np.shape(localmin)[1]
    
    R = localmin[1:row-1, 1:col-1] > localmin[1:row-1, 2:col]
    L = localmin[1:row-1, 1:col-1] > localmin[1:row-1, 0:col-2]
    U = localmin[1:row-1, 1:col-1] > localmin[0:row-2, 1:col-1]
    D = localmin[1:row-1, 1:col-1] > localmin[2:row, 1:col-1]
    UR = localmin[1:row-1, 1:col-1] > localmin[0:row-2, 2:col]
    DR = localmin[1:row-1, 1:col-1] > localmin[2:row, 2:col]
    UL = localmin[1:row-1, 1:col-1] > localmin[0:row-2, 0:col-2]
    DL = localmin[1:row-1, 1:col-1] > localmin[2:row, 0:col-2]
    
    m = np.zeros(sz)
    m[1:sz[0]-1, 1:sz[1]-1] = R * L * U * D * UR * DR * UL * DL
    localmin1 = m * localmin
    corners = np.zeros((nCorners, 2))
    p = localmin1.copy()
    findMax(nCorners,p,corners)
    return corners
```

### SSD Matching [1 pts]

Write a function ssd_match that implements the SSD matching algorithm for two input windows.

```python
def ssd_match(img1, img2, c1, c2, R):
    """Compute SSD given two windows.

    Args:
        img1: Image 1.
        img2: Image 2.
        c1: Center (in image coordinate) of the window in image 1.
        c2: Center (in image coordinate) of the window in image 2.
        R: R is the radius of the patch, 2 * R + 1 is the window size

    Returns:
        SSD matching score for two input windows.

    """
    matching_score = 0
    width = min(img1.shape[0],img2.shape[0])
    height = min(img1.shape[1],img2.shape[1])
    I1 = np.zeros(((2*R+1)**2,1))
    I2 = np.zeros(((2*R+1)**2,1))
    n = 0
    w = int(min(width,c1[1]+R+1))
    h = int(min(height,c1[0]+R+1))
    sw = int(max(0,c1[1]-R))
    sh = int(max(0,c1[0]-R))
    for i in range(sw,w):
        for j in range(sh,h):
            I1[n] = img1[i,j]
            n = n+1
    m = 0
    w = int(min(width,c2[1]+R+1))
    h = int(min(height,c2[0]+R+1))
    sw = int(max(0,c2[1]-R))
    sh = int(max(0,c2[0]-R))
    for i in range(sw,w):
        for j in range(sh,h):
            I2[m] = img2[i,j]
            m = m+1
    c = max(n,m)
    for i in range(c):
        diff = I1[i] - I2[i]
        matching_score += diff**2
    return matching_score
```

### Naive Matching [3 pts]

Equipped with the corner detector and the SSD matching function, we are ready to start finding correspondances. One naive strategy is to try and find the best match between the two sets of corner points. Write a script that does this, namely, for each corner in image1, find the best match from the detected corners in image2 (or, if the SSD match score is too low, then return no match for that point). You will have to figure out a good threshold (SSDth) value by experimentation. Write a function naiveCorrespondanceMatching.m and call it as below. Examine your results for 10, 20, and 30 detected corners in each image. Choose a number of detected corners to the maximize the number of correct matching pairs. naive_matching will call your SSD mathching code. <img src = "fig/dinoMatch.png", alt="dino match" width="800"/>

```python
def naive_matching(img1, img2, corners1, corners2, R, SSDth):
    """Compute SSD given two windows.

    Args:
        img1: Image 1.
        img2: Image 2.
        corners1: Corners in image 1 (nx2)
        corners2: Corners in image 2 (nx2)
        R: SSD matching radius
        SSDth: SSD matching score threshold

    Returns:
        SSD matching result a list of tuple (c1, c2), 
        c1 is the 1x2 corner location in image 1, 
        c2 is the 1x2 corner location in image 2. 
    """
    matching = []
    best_match = np.ones(np.shape(corners1)[0])
    best_match = best_match*SSDth
    for i in range(np.shape(corners1)[0]):
        best = -1
        for j in range(np.shape(corners2)[0]):
            ssd = ssd_match(img1, img2, corners1[i], corners2[j], R)
            if ssd < best_match[i]:
                best = j
                best_match[i] = ssd
        if best != -1:
            matching.append((corners1[i],corners2[best]))                
    return matching
```

```python
import numpy as np
from scipy.misc import imread
import matplotlib.pyplot as plt
from scipy.io import loadmat

def compute_fundamental(x1,x2):
    """    Computes the fundamental matrix from corresponding points 
        (x1,x2 3*n arrays) using the 8 point algorithm.
        Each row in the A matrix below is constructed as
        [x'*x, x'*y, x', y'*x, y'*y, y', x, y, 1] 
    """
    
    n = x1.shape[1]
    if x2.shape[1] != n:
        raise ValueError("Number of points don't match.")
    
    # build matrix for equations
    A = np.zeros((n,9))
    for i in range(n):
        A[i] = [x1[0,i]*x2[0,i], x1[0,i]*x2[1,i], x1[0,i]*x2[2,i],
                x1[1,i]*x2[0,i], x1[1,i]*x2[1,i], x1[1,i]*x2[2,i],
                x1[2,i]*x2[0,i], x1[2,i]*x2[1,i], x1[2,i]*x2[2,i] ]
            
    # compute linear least square solution
    U,S,V = np.linalg.svd(A)
    F = V[-1].reshape(3,3)
        
    # constrain F
    # make rank 2 by zeroing out last singular value
    U,S,V = np.linalg.svd(F)
    S[2] = 0
    F = np.dot(U,np.dot(np.diag(S),V))
    
    return F/F[2,2]


def fundamental_matrix(x1,x2):
    n = x1.shape[1]
    if x2.shape[1] != n:
        raise ValueError("Number of points don't match.")

    # normalize image coordinates
    x1 = x1 / x1[2]
    mean_1 = np.mean(x1[:2],axis=1)
    S1 = np.sqrt(2) / np.std(x1[:2])
    T1 = np.array([[S1,0,-S1*mean_1[0]],[0,S1,-S1*mean_1[1]],[0,0,1]])
    x1 = np.dot(T1,x1)
    
    x2 = x2 / x2[2]
    mean_2 = np.mean(x2[:2],axis=1)
    S2 = np.sqrt(2) / np.std(x2[:2])
    T2 = np.array([[S2,0,-S2*mean_2[0]],[0,S2,-S2*mean_2[1]],[0,0,1]])
    x2 = np.dot(T2,x2)

    # compute F with the normalized coordinates
    F = compute_fundamental(x1,x2)

    # reverse normalization
    F = np.dot(T1.T,np.dot(F,T2))

    return F/F[2,2]
```

```python
def lines(line,x,y):
    xmin = x[0]
    xmax = x[1]
    ymin = y[0]
    ymax = y[1]
    allPts = np.array([[xmin, -(np.dot(line[0],xmin) + line[2])/line[1]],
                      [xmax, -(np.dot(line[0],xmax) + line[2])/line[1]],
                      [-(np.dot(line[1],ymin) + line[2])/line[0], ymin],
                      [-(np.dot(line[1],ymax) + line[2])/line[0], ymax]])
    count = 0
    pts =np.zeros((2,2))
    for i in range(4):
        if ((allPts[i,0] >= xmin) and (allPts[i,0] <= xmax) and 
        (allPts[i,1] >= ymin) and (allPts[i,1] <= ymax)):
            count = count + 1
            if (count == 1):
                pts[count-1,:] = allPts[i,:]
            elif (count > 1):
                addPoint = True
                for j in range(count-1):
                    diff = sum(abs(pts[j,:] - allPts[i,:]))
                    if (diff < 1e-5):
                        addPoint = False  
                if not addPoint:
                    count = count - 1
                elif (count > 2): print "error"
                else: pts[count-1,:] = allPts[i,:]
    return pts               
```

```python
def plot_epipolar_lines(img1,img2, cor1, cor2):
    """Plot epipolar lines on image given image, corners

    Args:
        img1: Image 1.
        img2: Image 2.
        cor1: Corners in homogeneous image coordinate in image 1 (3xn)
        cor2: Corners in homogeneous image coordinate in image 2 (3xn)

    """
    sz1 = np.shape(img1)
    xx=np.array([0,sz1[0]-1])
    yy=np.array([0,sz1[1]-1])
    F = fundamental_matrix(cor1, cor2)
    n = np.shape(cor1)[1]
    c2 = np.zeros(np.shape(cor2))
    plt.imshow(img1)
    pts2 = np.zeros((2,2,n))
    for i in range(n):
        plt.scatter(cor1[0,i], cor1[1,i], s=35, edgecolors='b', facecolors='b')
        c2[:,i]=np.dot(F,cor2[:,i])
        pts2[:,:,i]=lines(c2[:,i].T,yy,xx)
    for i in range(n):
        plt.plot([pts2[0,0,i],pts2[1,0,i]],[pts2[0,1,i],pts2[1,1,i]],linewidth = '1',color = 'b')
    plt.show()

    sz2 = np.shape(img2)
    xxx=np.array([0,sz2[0]-1])
    yyy=np.array([0,sz2[1]-1])
    c1 = np.zeros(np.shape(cor1))
    plt.imshow(img2)
    m = np.shape(cor2)[1]
    pts1 = np.zeros((2,2,m))
    for i in range(m):
        plt.scatter(cor2[0,i], cor2[1,i], s=35, edgecolors='b', facecolors='b')
        c1[:,i]=np.dot(F.T,cor1[:,i])
        pts1[:,:,i]=lines(c1[:,i].T,yyy,xxx)
    for i in range(m):
        plt.plot([pts1[0,0,i],pts1[1,0,i]],[pts1[0,1,i],pts1[1,1,i]],linewidth = '1',color = 'b')
    plt.show()
```

```python
from numpy import linalg as la 
def triangulate(corsSSD, P1, P2):
    """Find corner correspondence along epipolar line.

    Args:
        corsSSD: Corner correspondence
        P1: Projection matrix 1
        P2: Projection matrix 2
    
    Returns:
        3D points from triangulation
        
    """
    points3D = []
    for p1,p2 in corsSSD:
        I1x = p2[0]
        I1y = p2[1]
        I2x = p1[0]
        I2y = p1[1]
        A = np.vstack((I1x*P1[2]-P1[0],I1y*P1[2]-P1[1],I2x*P2[2]-P2[0],I2y*P2[2]-P2[1]))
        U,S,V = la.svd(A)
        points3D.append(V[-1])
    return points3D
```

```python
def find_outliers(points3D, P2, outlierTH, corsSSD):
    """Find and plot outliers on image 2

    Args:
        points3D: 3D point from triangulation
        P2: Projection matrix in for image 2
        outlierTH: outlier pixel distance threshold
        corsSSD: Corner correspondence
        
    """
    inlier = []
    outlier = []
    original = []
    for p,(p2,p1) in zip(points3D,corsSSD):
        p_homo = np.dot(P2,p)
        p_2d = (p_homo/p_homo[-1])[:-1]
        if p_2d[0]>0 and p_2d[1]>0:
            diff = sum(np.sqrt((p2-p_2d)**2))
            if diff <= outlierTH:
                inlier.append(p_2d)
            else:
                outlier.append(p_2d)
        if p2[0]>0 and p2[1]>0:
            original.append(p2)
    return np.array(original).T,np.array(inlier).T,np.array(outlier).T
```
