```python
import numpy as np
from scipy.misc import imread
import matplotlib.pyplot as plt
from scipy.ndimage.filters import gaussian_filter
from numpy import linalg as la
```

```python
def rgb2gray(rgb):
    """ Convert rgb image to grayscale.
    """
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
windowSize = 3
Ixx = np.zeros(Ix.shape)
Ixy = np.zeros(Ix.shape)
Iyy = np.zeros(Ix.shape)
lam_1 = np.zeros(Ix.shape)
lam_2 = np.zeros(Ix.shape)
e = np.zeros(Ix.shape)
for x in xrange(0, 1200-windowSize):
    for y in xrange(0, 1200-windowSize):
        for i in range(x,x+windowSize):
            for j in range(y,y+windowSize):
                Ixx[x,y] += Ix[i,j]*Ix[i,j]
                Iyy[x,y] += Iy[i,j]*Iy[i,j]
                Ixy[x,y] += Ix[i,j]*Iy[i,j]
        detM = (Ixx[x,y]* Iyy[x,y]) - (Ixy[x,y]* Ixy[x,y])
        traceM = Ixx[x,y] + Iyy[x,y] #lam1 + lam2
        S = np.sqrt(traceM*traceM-4*detM) #lam1 - lam2
        lam_1[x,y] = (traceM+S)/2
        lam_2[x,y] = (traceM-S)/2
        e[x,y] = min(lam_1[x,y],lam_2[x,y])
```
